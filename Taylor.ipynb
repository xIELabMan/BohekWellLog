```python
# Import tools

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import bokeh as bk


from pandas import Series, DataFrame
from matplotlib.pyplot import figure

from bokeh.plotting import figure, output_file, show, ColumnDataSource
from bokeh.layouts import column, row
from bokeh.models import Range1d, HoverTool, LinearAxis, Legend

plt.style.use('ggplot')
```

```python
# Read CSV, preview data
dfB = pd.read_csv('taylor.txt', sep='\t', header=0)
maxdepth = float(min(dfB['DEPTH (FT)']))
mindepth = float(max(dfB['DEPTH (FT)']))
# dfB
```

```python
depth=dfB['DEPTH (FT)']

#Track 1 Variables
CALI = dfB['CALI (IN)']
GR = dfB['GR (GAPI)']

#Track 2 Variables
NPHISS = dfB['NPHISS']
RHOB = dfB['RHOB (G/C3)']

#Track 3 Variables
RILD = dfB['RILD (OHMM)']
SP = dfB['SP (MV)']
RFOC = dfB['RFOC (OHMM)']

ymax = float(min(depth))
ymin = float(max(depth))
```

```python
# #https://bokeh.pydata.org/en/latest/docs/user_guide/tools.html#basic-tooltips

TOOLTIPS1 = HoverTool(tooltips =[
        ( 'depth',  '$y'   ),
        ( 'CALI',   '@CALI (IN)'   ), # use @{ } for field names with spaces
        ( 'GR',   '$x'     )], 
        mode = 'hline')
```

```python
#https://bokeh.pydata.org/en/latest/docs/user_guide/tools.html#basic-tooltips

# TOOLTIPS1 = HoverTool()
# TOOLTIPS1.tooltips =([('(depth, LCAL, HSGR, HCGR)', '($y, $LCAL, $HSGR, $HCGR)')])
```

```python
output_file("taylor.html")

# Tracks with four traces
# https://bokeh.pydata.org/en/latest/docs/user_guide/interaction/linking.html
t1 = bk.plotting.figure(title='CALI (IN), GR (GAPI)',
            extra_x_ranges={'CALI (IN)': Range1d(start=0, end=16),
                           'GR (GAPI)': Range1d(start=0, end=250),
                           },
            plot_width=250, plot_height=750,
            y_range=(ymin, ymax), 
            x_axis_location='above'
            )
t1.tools.append(TOOLTIPS1)
t1.extra_x_ranges = {'CALI (IN)': Range1d(start=0, end=16)}
t1.add_layout(LinearAxis(x_range_name='CALI (IN)'), 'above')


t1.line(x=dfB['GR (GAPI)'], y=dfB['DEPTH (FT)'], color='red')
t1.line(x=dfB['CALI (IN)'], y=dfB['DEPTH (FT)'], x_range_name='CALI (IN)', color='black')
#Track 2
t2 = bk.plotting.figure(title='NPHISS, RHOB (G/C3)',
            plot_width=250, plot_height=750,
            y_range=t1.y_range, 
            x_axis_location='above'
            )

t2.axis.axis_label_standoff=100
t2.line(x=dfB['NPHISS'], y=dfB['DEPTH (FT)'], color='red')
t2.line(x=dfB['RHOB (G/C3)'], y=dfB['DEPTH (FT)'], color='blue')

# #Track 3
t3 = bk.plotting.figure(title='RILD (OHMM), SP (MV), RFOC (OHMM)', 
            plot_width=250, plot_height=750,
            y_range=t1.y_range, 
            x_axis_location='above')

t3.multi_line(xs=[dfB['RILD (OHMM)'], dfB['SP (MV)'], dfB['RFOC (OHMM)']], ys=[dfB['DEPTH (FT)'], dfB['DEPTH (FT)'], dfB['DEPTH (FT)']], color=['black','green','yellow'])

# #Track 4
# t4 = bk.plotting.figure(title='RM_HRLT, RT_HRLT, DTCO, DTST',
#             plot_width=250, plot_height=750,
#             y_range=t1.y_range, 
#             x_axis_location='above')
# t4.multi_line(xs=[dfB['RM_HRLT'], dfB['RT_HRLT'], dfB['DTCO'], dfB['DTST']], ys=[dfB['depth'], dfB['depth'], dfB['depth'], dfB['depth']], color=['black','blue','red','green'])
# show(row(t1, t2, t3, t4))

show(row(t1, t2, t3))

```

```python
output_file("taylor.html")
t1 = bk.plotting.figure(title='CALI (IN), GR (GAPI)',
            plot_width=250, plot_height=750,
            y_range=(ymin, ymax), 
            x_axis_location='above'
            )

t1.extra_x_ranges = {'CALI (IN)': Range1d(start=4, end=16)}
t1.add_layout(LinearAxis(x_range_name='CALI (IN)'), 'above')

t1.line(x=dfB['GR (GAPI)'], y=dfB['DEPTH (FT)'], color='red')
t1.line(x=dfB['CALI (IN)'], y=dfB['DEPTH (FT)'], x_range_name='CALI (IN)', color='black')


t2 = bk.plotting.figure(title='NPHISS, RHOB (G/C3)',
            plot_width=250, plot_height=750,
            y_range=(ymin, ymax), 
            x_axis_location='above'
            )

t2.axis.axis_label_standoff=100
t2.line(x=dfB['NPHISS'], y=dfB['DEPTH (FT)'], color='red')
t2.line(x=dfB['RHOB (G/C3)'], y=dfB['DEPTH (FT)'], color='blue')

t3 = bk.plotting.figure(title='RILD (OHMM), SP (MV), RFOC (OHMM)', 
            plot_width=250, plot_height=750,
            y_range=(ymin, ymax), 
            x_axis_location='above')

t3.line(dfB['SP (MV)'], y=dfB['DEPTH (FT)'], color='green')
t3.line(x=dfB['RILD (OHMM)'], y=dfB['DEPTH (FT)'], color='black')
t3.line(dfB['RFOC (OHMM)'], dfB['DEPTH (FT)'], color='orange')

# t1.tools.append(TOOLTIPS1)
show(row(t1,t2,t3))
```
